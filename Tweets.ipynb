## I am using a dataset from the internet, for demo and have also created slur.txt for demo basis.

### Importing the Required Modules

```python
import pandas as pd
import numpy as np
import re
```

```python
# I am assuming here that tweets.csv is the file which has all the tweets of the users
df = pd.read_csv('tweets.csv', encoding='utf-8')
```

```python
# dropping the label column, because it is same as id
df.drop('label',axis=1,inplace=True)
```

```python
df
```

## Cleaning the Tweets

### Function for removing all the useless words:

```python
# The function returns the same input string but without the given pattern
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt
```

```python
# remove twitter handles (@user)
df['tidy_tweet'] = np.vectorize(remove_pattern)(df['tweet'], "@[\w]*")
```

```python
df
```

```python
# remove special characters, numbers, punctuations
df['tidy_tweet'] = df['tidy_tweet'].str.replace("[^a-zA-Z]", " ") 
```

```python
df
```

```python
# Removing all the words with characters less than to 3
df['tidy_tweet'] = df['tidy_tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>=3]))
```

```python
df
```

```python
# read file with slur words
with open('slurs.txt', 'r', encoding='utf-8') as f:
    racial_slurs = f.read().splitlines()

pattern = re.compile('|'.join(racial_slurs), re.IGNORECASE)
```

```python
pattern
```

## Finding the degree of Porfanity

```python
# iterate over tweets and check for racial slurs
for tweet in df['tidy_tweet']:
    # count number of racial slurs in tweet
    num_slurs = len(re.findall(pattern, tweet))
    
    # calculate degree of profanity as a percentage of total words in tweet
    if len(tweet.split()) > 0: # checking for divide by zero error
        degree_of_profanity = num_slurs / len(tweet.split()) * 100
        
    # print degree of profanity for tweet
    print(f"Tweet: {tweet.strip()} \nDegree of Profanity: {degree_of_profanity:.2f}%\n")
```
